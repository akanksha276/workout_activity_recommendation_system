```python
# Content-Based Filtering
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
df['sport'] = df['sport'].fillna('')  # Handle missing values
tfidf_matrix = tfidf_vectorizer.fit_transform(df['sport'])

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get content-based recommendations
def get_content_based_recommendations(activity_name):
    idx = df[df['id'] == activity_name].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Top 10 similar activities
    activity_indices = [i[0] for i in sim_scores]
    return df['id'].iloc[activity_indices]

# Example: Get content-based recommendations for an activity
activity_id = 12345  # Replace with an actual activity ID
content_based_recommendations = get_content_based_recommendations(activity_id)
print("Content-Based Recommendations:", content_based_recommendations.tolist())
```

```python
from lightfm import LightFM
from lightfm.data import Dataset
from lightfm.evaluation import precision_at_k
from scipy.sparse import coo_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_encoded = pd.get_dummies(filtered_df, columns=['gender', 'sport'],dtype=int)
```

```python
df_encoded.head(10)
```

```python
df_encoded.info()
```

```python
# df_encoded.drop(['id','heart_rate','derived_speed','since_begin','since_last'],axis=1,inplace=True)
```

```python
train_data, test_data = train_test_split(df_encoded, test_size=0.2, random_state=42)
```

```python
# Create a user-item interaction matrix
dataset_train = Dataset()
dataset_train.fit(users=train_data['userId'], items=train_data.columns)
(interactions_train, _) = dataset_train.build_interactions([(user, item, 1) for user, item in zip(train_data['userId'], train_data.columns)])
```

```python
# Create item feature matrix
item_features_train = coo_matrix(train_data.drop(['userId'], axis=1).values)
```

```python
print('working')
```

```python
# Build the model with content-based features
model = LightFM(loss='warp', random_state=42)
model.fit(interactions_train, item_features=item_features_train, epochs=10, num_threads=2)
```

```python
dataset_test = Dataset()
dataset_test.fit(users=test_data['userId'], items=test_data.columns)
(interactions_test, _) = dataset_test.build_interactions([(user, item, 1) for user, item in zip(test_data['userId'], test_data.columns)])
```

```python
item_features_test = coo_matrix(test_data.drop(['userId'], axis=1).values)
```

```python
test_precision = precision_at_k(model, interactions_test, k=5, item_features=item_features_test).mean()
```

```python
test_precision
```

```python
# Example: Get recommendations for a user from the test set
user_id_to_recommend = test_data['userId'].iloc[0]
user_index = dataset_test.mapping()[0][user_id_to_recommend]
# Convert the range object to a list
item_indices = list(range(len(test_data.columns)))
```

```python
# Predict scores for items
scores = model.predict(user_index, item_indices, item_features=item_features_test)
```

```python
# Get indices of top recommended activities
top_indices = scores.argsort()[::-1]

# Get the corresponding activity names
recommendations = test_data.columns[top_indices].tolist()

print(f"Recommendations for User {user_id_to_recommend} in the test set: {recommendations}")
print(f"Precision at K on the test set: {test_precision}")
```
